```python
class JungleMonkey():
    def __init__(self):
        print("\t\t\t\t The Jungle Monkey Multicuisine Restuarant ")
        print("\t\t\t\t\t ESTD.2024")
        print("\t\t\t\t Bringing class to cuisine")
        print("MENU")
        
        x = ["Indian", "Chinese", "Drinks"]
        for k in x:
            print("\n\n", k)

    def Indian(self):
        x = ["Pav Bhaji", "Chole Bhature", "Masala Dosa", "Idli Sambhar", "Paneer Tikka", "Biryani", "Samosa chaat", "Dahi Papdi"]
        y = [150, 180, 200, 130, 125, 250, 210, 90]
        df = {"Indian Food Types": x, "Price": y}
        import pandas as pd
        df = pd.DataFrame(df)
        print("\n\n Indian Foods \n")
        print(df)
        
        d = []
        while True:
            b = int(input("\n\nEnter the cuisine position : "))
            c = int(input("Enter the quantity: "))
            z = df["Price"][b]
            cost = z * c
            d.append(cost)
            
            reorder = input("You want to reorder (yes/no): ")
            if reorder == "no":
                print(f"Thank you, Please visit again")
                break
        
        total = sum(d)
        service_tax = total * 0.05
        total_with_tax = total + service_tax
        
        from colorama import Fore, Back, Style
        print(Fore.BLACK + Back.YELLOW + Style.BRIGHT + "Service tax = 5%")
        print("Total Amount:", total, "Rupees")
        print("Total Amount Including Service Tax:", total_with_tax, "Rupees")

    def Chinese(self):
        x = ["Hakka Noodles", "Manchurian", "Chilli Paneer", "Spring rolls", "Mushroom chilli", "Veg Momos", "Sweet tofu", "Schezwan fried rice"]
        y = [150, 180, 200, 130, 125, 250, 210, 90]
        df = {"Chinese Food Types": x, "Price": y}
        import pandas as pd
        df = pd.DataFrame(df)
        print("\n\n Chinese Foods \n")
        print(df)
        
        d = []
        while True:
            b = int(input("Enter the cuisine position : "))
            c = int(input("Enter the quantity: "))
            z = df["Price"][b]
            cost = z * c
            d.append(cost)
            
            reorder = input("You want to reorder (yes/no): ")
            if reorder == "no":
                print(f"Thank you, Please visit again")
                break
        
        total = sum(d)
        service_tax = total * 0.05
        total_with_tax = total + service_tax
        
        from colorama import Fore, Back, Style
        print(Fore.BLACK + Back.YELLOW + Style.BRIGHT + "Service tax = 5%")
        print("Total Amount:", total, "Rupees")
        print("Total Amount Including Service Tax:", total_with_tax, "Rupees")

    def drinks(self):
        x = ["Masala Chai", "Cold Coffee", "Espresso", "Cafe Latte", "Blue lagoon", "Green tea", "Mix fruit smoothie", "Lemon Iced Tea"]
        y = [30, 180, 120, 160, 195, 80, 60, 210]
        df = {"Drink Types": x, "Price": y}
        import pandas as pd
        df = pd.DataFrame(df)
        print("\n\n Drinks Menu \n")
        print(df)
        
        d = []
        while True:
            b = int(input("Enter the cuisine position: "))
            c = int(input("Enter the quantity: "))
            z = df["Price"][b]
            cost = z * c
            d.append(cost)
            
            reorder = input("You want to reorder (yes/no): ")
            if reorder == "no":
                print(f"Thank you, Please visit again")
                break
        
        total = sum(d)
        service_tax = total * 0.05
        total_with_tax = total + service_tax
        
        from colorama import Fore, Back, Style
        print(Fore.BLACK + Back.YELLOW + Style.BRIGHT + "Service tax = 5%")
        print("Total Amount:", total, "Rupees")
        print("Total Amount Including Service Tax:", total_with_tax, "Rupees")

obj = JungleMonkey()
ft = input("Enter the Food type (Indian/Chinese/Drinks): ").capitalize()

if ft == "Indian":
    obj.Indian()
elif ft == "Chinese":
    obj.Chinese()
elif ft == "Drinks":
    obj.drinks()
else:
    print("Invalid food type.")
```
